```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/drive/MyDrive/IPL_Matches_2008_2022.csv")
```

```python
data
```

```python
sns.set_style("darkgrid")
wins = pd.DataFrame(data['WinningTeam'].value_counts())
wins['name'] = wins.index
plt.figure(facecolor='#D3D3D3')
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=12)
plt.xlabel("Teams")
plt.ylabel("No.of.Wins")
plt.title('Total wins by each team', fontsize=20)
ax = sns.barplot(x='name', y='WinningTeam', data=wins, hue='name', color='gray')

for container in ax.containers:
    ax.bar_label(container, fontsize=10)

plt.show()
```

```python
players=pd.DataFrame(data['Player_of_Match'].value_counts())
players['name']=players.index
players=players.head(20)
plt.figure(facecolor='#d9ffcc')
plt.xticks(fontsize=10)
plt.yticks(fontsize=12)
plt.xlabel("No.of.wins")
plt.ylabel("Players")
plt.title('Top_20 Players with most "Man of the match" Awards',fontsize=20)
palette=sns.color_palette("deep",len(players))
ax=sns.barplot(x='Player_of_Match',y='name',data=players,hue='name',color="green")
for i in ax.containers:
    ax.bar_label(i,fontsize=10)
```

```python
mi_csk_rcb_data = data[(data['Team1'] == 'Mumbai Indians') | (data['Team1'] == 'Chennai Super Kings') | (data['Team1'] == 'Royal Challengers Bangalore')]
team_wins = mi_csk_rcb_data.groupby('Team1')['WinningTeam'].value_counts().reset_index(name='count')
top_10_teams = team_wins.sort_values(by='count', ascending=False).head(10)
plt.figure(figsize=(10, 10),facecolor="#D3D3D3")
plt.pie(top_10_teams['count'], labels=top_10_teams['Team1'], autopct='%1.1f%%')
plt.title('Top 10 Teams by Wins (MI, CSK, RCB)')
plt.show()
```
